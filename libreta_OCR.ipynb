Si el PDF deseado ya está guardado en "pdf_folder", usaremos la función "pdf_a_jpg_y_guardar"

Esto convertirá cada página del PDF en una imagen que podremos manipular para después extraer el texto

La función ocupa varios parámetro opcionales y uno obligatorio: "pdf_nombre", que deberá ser, entre comillas y con extensión, el nombre del PDF

```python
pdf_a_jpg_y_guardar(pdf_nombre="logica_intro.pdf")
```

El folder "imagenes_convertidas_a_jpg" deberá tener ahora una imagen por cada página en el PDF

Ahora describiré el proceso para una página individual, aunque al final habrá código para extraer el texto de todas las páginas a la vez

Haremos que las página elegida esté en blanco y negro con "jpg_a_byn_y_guardar"

La función ocupa varios parámetros opcionales y uno obligatorio; "numero_de_pagina", que deberá ser el número con que se guardó la página deseada

```python
jpg_a_byn_y_guardar(numero_de_pagina=1)
```

El folder "imagenes_a_byn" deberá tener ahora la imagen en blanco y negro, con la palabra "gris" añadida al nombre

Finalmente extremos el texto de la imagen a blanco y negro con la función "obtener_texto"

La función ocupa varios parámetros opcionales y uno obligatorio: "numero_de_pagina"

Por default la función asume que el texto está en español; si está en otro idioma hay que especificarlo con el parámetro "idioma"

Opciones de idioma incluyen "spa"=español, "eng"=inglés, "fra"=francés, "ita"=italiano, "deu"=alemán

```python
obtener_texto(numero_de_pagina=1)
```

A continuación el código para convertir en blanco y negro de todas las páginas:

Alterar el valor de "numero_de_paginas"

```python
numero_de_paginas = 5

for i in range(numero_de_paginas):
    jpg_a_byn_y_guardar(numero_de_pagina=i+1)
```

Y finalmente el código para extraer el texto de todas las páginas:

Alterar el valor de "numero_de_paginas"

```python
numero_de_paginas = 5

for i in range(numero_de_paginas):
    print(f"PÁGINA {i+1}:\n\n")
    obtener_texto(numero_de_pagina=i+1)
    print("")
```
